```python
cleaned_restaurant_df = pd.read_csv("NYC-cleaned(in).csv")
cleaned_restaurant_df
```

```python
cleaned_rat_df = cleaned_rat_df[['Unique Key','Created Date','Closed Date','Agency','Agency Name','Complaint Type','Descriptor','Location Type','Incident Zip','Incident Address','Street Name','Address Type','City','Status','Due Date','Resolution Action Updated Date','Community Board','Borough','Location']]
```

```python
cleaned_rat_df
```

```python
cleaned_restaurant_df.isnull().sum()
```

```python
cleaned_restaurant_df = cleaned_restaurant_df[['CAMIS','DBA','BORO','BUILDING','STREET','ZIPCODE','PHONE','CUISINE DESCRIPTION','INSPECTION DATE','ACTION','VIOLATION CODE','VIOLATION DESCRIPTION','CRITICAL FLAG','SCORE','RECORD DATE','INSPECTION TYPE']]
```

```python
cleaned_restaurant_df
```

```python
df_rat = cleaned_rat_df.dropna()
```

```python
df_rat
```

```python
df_restaurant = cleaned_restaurant_df.dropna()
```

```python
df_restaurant
```

```python
df_rat.dtypes
```

```python
df_restaurant.dtypes
```

```python
restaurant_grouped = (
    df_restaurant
        .groupby("ZIPCODE")
        .agg(
            total_inspections=("CAMIS", "count"),
            unique_restaurants=("CAMIS", "nunique"),
            avg_score=("SCORE", "mean"),
            median_score=("SCORE", "median"),
            grade_A_pct=("ACTION", lambda x: (x.str.contains("Grade A", na=False)).mean()),
        )
        .reset_index()
)
```

```python
rat_grouped = (
    df_rat
        .groupby("Incident Zip")
        .agg(
            rat_sightings=("Unique Key", "count")
        )
        .reset_index()
)
```

```python
merged_df = rat_grouped.merge(
    restaurant_grouped,
    left_on="Incident Zip",
    right_on="ZIPCODE",
    how="left"
)
```

```python
merged_df = merged_df.fillna(0)
```

```python
merged_df
```

```python
df_rat.to_csv("cleaned_rat.csv", index=False)
```

```python
df_restaurant.to_csv("cleaned_restaurant.csv", index=False)
```

```python
merged_df.to_csv("merged_df.csv", index=False)
```
